## Q2

```python
array=np.array([10, 52, 62, 16, 16, 54, 453])
print("Indices of Sorted array",array.argsort())
array.sort()
print("Sorted array: ",array)
print("Four smallest elements",array[:4 ])
print("Five largest elements",array[-5:])
```

```python
array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])
int_elements = array[array % 1 == 0]
print("Integer elements only: ",int_elements)
float_elements=array[array % 1 !=0]
print("Float elements only",float_elements)
```

```python
x=dataset[0]
tax_rate=((x % 5) + 5) / 100
taxed_sales=tax_rate*dataset
print("Sales Tax for Each Value:",taxed_sales)
```
