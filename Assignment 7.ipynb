```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Q1

**P I**

```python
np.random.seed(102317032)

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']

sales_data = np.random.randint(1000, 5001, size=(12, 4))

sales_df = pd.DataFrame(sales_data, columns=categories, index=months)

print(sales_df)
```

**Part II**

```python
print(sales_df.head())
print(sales_df.describe())
```

```python
category_totals = sales_df.sum()
monthly_totals = sales_df.sum(axis=1)
print("Total sales per category:")
print(category_totals)
print("\nTotal sales per month:")
print(monthly_totals)
```

```python
sales_growth = sales_df.pct_change().mean()
print("\nAverage sales growth per category:")
print(sales_growth)
```

```python
sales_df['Total Sales'] = monthly_totals
sales_df['Growth Rate'] = sales_df['Total Sales'].pct_change() * 100
```

```python
roll_number = int(input("Enter your roll number: "))
if roll_number % 2 == 0:
    sales_df['Electronics'] = sales_df['Electronics'] * 0.9  # 10% discount for even roll no.
else:
    sales_df['Clothing'] = sales_df['Clothing'] * 0.85  # 15% discount for odd roll no.
```

```python
print(sales_df)
```

**Part III**

```python
plt.figure(figsize=(10, 5))
for category in categories:
    plt.plot(sales_df.index, sales_df[category], marker='o', label=category)
plt.title("Monthly Sales Trends")
plt.xlabel("Month")
plt.ylabel("Sales Units")
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(8, 5))
sns.boxplot(data=sales_df[categories])
plt.title("Sales Distribution by Category")
plt.xlabel("Product Categories")
plt.ylabel("Sales Units")
plt.show()
```

## Q2

```python
array = np.array([[1, -2, 3], [-4, 5, -6]])

abs_array = np.abs(array)
print("\nElement-wise absolute values:")
print(abs_array)

percentiles_flat = np.percentile(array, [25, 50, 75])
percentiles_col = np.percentile(array, [25, 50, 75], axis=0)
percentiles_row = np.percentile(array, [25, 50, 75], axis=1)
print("\nPercentiles for flattened array (25th, 50th, 75th):", percentiles_flat)
print("\nPercentiles for each column:")
print(percentiles_col)
print("\nPercentiles for each row:")
print(percentiles_row)

mean_flat = np.mean(array)
median_flat = np.median(array)
std_flat = np.std(array)

mean_col = np.mean(array, axis=0)
median_col = np.median(array, axis=0)
std_col = np.std(array, axis=0)

mean_row = np.mean(array, axis=1)
median_row = np.median(array, axis=1)
std_row = np.std(array, axis=1)

print("\nMean, Median, Std of flattened array:")
print(mean_flat, median_flat, std_flat)

print("\nMean, Median, Std for each column:")
print(mean_col, median_col, std_col)

print("\nMean, Median, Std for each row:")
print(mean_row, median_row, std_row)
```

```python
import numpy as np

a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

floor_values = np.floor(a)       
ceiling_values = np.ceil(a)      
truncated_values = np.trunc(a)   
rounded_values = np.round(a)    

print("Original Array:       ", a)
print("Floor Values:         ", floor_values)
print("Ceiling Values:       ", ceiling_values)
print("Truncated Values:     ", truncated_values)
print("Rounded Values:       ", rounded_values)
```

## Q4

```python
def swap_elements(lst, index1, index2):
    temp = lst[index1]  
    lst[index1] = lst[index2] 
    lst[index2] = temp 

my_list = [10, 20, 30, 40, 50]
print("Original List:", my_list)

swap_elements(my_list, 1, 3)

print("List after swapping:", my_list)
```

## Q5

```python
my_set = {10, 20, 30, 40, 50}
print("Original Set:", my_set)

lst = list(my_set)

i, j = lst.index(20), lst.index(40)
lst[i], lst[j] = lst[j], lst[i]

my_set = set(lst)
print("Set after swapping:", my_set)
```
